# 3.8 다른 정형 데이터와의 결합

```python
import numpy as np
import pandas as pd

#data load
train = pd.read_csv('Desktop/multi_table_train.csv')
product_master = pd.read_csv('Desktop/multi_table_product.csv')
user_log = pd.read_csv('Desktop/multi_table_log.csv')
```

```python
train
```

- train: 학습 데이터(사용자 ID, 상품 ID, 목적변수 등의 열이 있음)

```python
product_master
```

- product_master: 상품 마스터(상품 ID와 상품의 정보를 나타내는 열이 있음)

```python
user_log
```

- user_log: 사용자 행동의 로그 데이터(사용자 ID와 각 행동의 정보를 나타내는 열이 있음)

```python
#train data와 product_master의 결합
train = train.merge(product_master, on='product_id', how='left')
```

```python
train.head()
```

```python
#user_log의 사용자별 행의 수를 구함
user_log_agg = user_log.groupby('user_id').size().reset_index().rename(columns={0: 'user_count'})
```

```python
user_log_agg.head()
```

```python
#위에서 구한 user_log_agg와 train data를 결합
train = train.merge(user_log_agg, on='user_id', how='left')
```
